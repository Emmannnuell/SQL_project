Tarea 3.1: Número de libros publicados después del 1 de enero de 2000
El objetivo de esta consulta es medir qué tan "moderno" o actualizado está el catálogo de la plataforma.

```python
# Consulta SQL para contar libros del nuevo milenio
query_1 = """
SELECT COUNT(*) AS total_books
FROM books
WHERE publication_date > '2000-01-01';
"""

# Ejecutar la consulta en nuestro motor local
df_resultado_1 = pd.io.sql.read_sql(query_1, con=engine)
df_resultado_1
```

Tarea 3.2: Número de reseñas y calificación promedio para cada libro
El objetivo de negocio: Queremos saber qué libros jalan más gente (volumen de reseñas) y cuáles son los mejor calificados (promedio de estrellas). Esto ayuda a identificar los "bestsellers" de la plataforma.

```python
# Consulta avanzada con CTEs para evitar duplicación de agregaciones
query_2 = """
WITH con_promedios AS (
    SELECT book_id, 
           AVG(rating) AS rating_avg
    FROM ratings
    GROUP BY book_id
),
con_reseñas AS (
    SELECT book_id, 
           COUNT(review_id) AS total_reviews
    FROM reviews
    GROUP BY book_id
)
SELECT b.book_id,
       b.title,
       COALESCE(p.rating_avg, 0) AS rating_avg,
       COALESCE(r.total_reviews, 0) AS total_reviews
FROM books b
LEFT JOIN con_promedios p ON b.book_id = p.book_id
LEFT JOIN con_reseñas r ON b.book_id = r.book_id;
"""

# Ejecutamos en la base local
df_resultado_2 = pd.io.sql.read_sql(query_2, con=engine)
df_resultado_2
```

Tarea 3.3: Editorial con el mayor número de libros de más de 50 páginas

```python
# Consulta para encontrar la editorial líder en contenido relevante
query_3 = """
SELECT p.publisher,
       COUNT(b.book_id) AS total_books
FROM books b
JOIN publishers p ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY total_books DESC
LIMIT 1;
"""

# Ejecutamos en la base local
df_resultado_3 = pd.io.sql.read_sql(query_3, con=engine)
df_resultado_3
```

Tarea 3.4: Autor con la calificación promedio más alta

```python
# Consulta de tres tablas con filtro condicional HAVING
query_4 = """
SELECT a.author,
       AVG(r.rating) AS avg_rating,
       COUNT(r.rating) AS total_ratings
FROM books b
JOIN ratings r ON b.book_id = r.book_id
JOIN authors a ON b.author_id = a.author_id
GROUP BY a.author
HAVING COUNT(r.rating) >= 2
ORDER BY avg_rating DESC
LIMIT 1;
"""

# Ejecutamos en la base local
df_resultado_4 = pd.io.sql.read_sql(query_4, con=engine)
df_resultado_4
```

Tarea 3.5: Número promedio de reseñas de texto entre los usuarios más activos

```python
# Consulta avanzada para promediar la actividad de los Power Users
query_5 = """
WITH user_stats AS (
    SELECT r.username,
           COUNT(DISTINCT r.book_id) AS total_ratings,
           COUNT(DISTINCT rev.review_id) AS total_reviews
    FROM ratings r
    LEFT JOIN reviews rev ON r.username = rev.username
    GROUP BY r.username
    HAVING COUNT(DISTINCT r.book_id) > 1
)
SELECT AVG(total_reviews) AS avg_text_reviews
FROM user_stats;
"""

# Ejecutamos en la base local
df_resultado_5 = pd.io.sql.read_sql(query_5, con=engine)
df_resultado_5
```
