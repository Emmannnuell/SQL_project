###  Nota sobre el Entorno de Ejecución (Base de Datos Local)
Para garantizar que este cuaderno sea 100% ejecutable, reproducible y autónomo, 
el proyecto migró de un servidor PostgreSQL remoto a una base de datos relacional local 
utilizando **SQLite**. Los datos de muestra incluidos replican de forma idéntica la estructura 
y las relaciones del modelo original, permitiendo que todas las consultas SQL se ejecuten 
en tiempo real sin depender de conexiones externas.

```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Crear el motor de conexión local
engine = create_engine('sqlite:///libros_local.db')

# 2. Definir datos de muestra 
authors_data = pd.DataFrame({
    'author_id': [1, 2, 3, 4],
    'author': ['Diana Gabaldon', 'Agatha Christie', 'Alan Moore', 'Albert Camus']
})

publishers_data = pd.DataFrame({
    'publisher_id': [1, 2, 3],
    'publisher': ['Penguin Books', 'Ace Books', 'Aladdin']
})

# Agregamos fechas anteriores y posteriores al 2000
books_data = pd.DataFrame({
    'book_id': [1, 2, 3, 4, 5],
    'author_id': [1, 2, 3, 4, 1],
    'title': ['Outlander', 'Murder on the Orient Express', 'Watchmen', 'The Stranger', 'Dragonfly in Amber'],
    'num_pages': [850, 256, 416, 150, 743],
    'publication_date': ['1991-06-01', '2002-03-05', '1987-09-01', '1942-05-19', '2005-08-02'],
    'publisher_id': [1, 1, 2, 3, 1]
})

# Calificaciones para evaluar promedios y recuentos
ratings_data = pd.DataFrame({
    'rating_id': [1, 2, 3, 4, 5, 6, 7],
    'book_id': [1, 1, 2, 2, 3, 5, 5],
    'username': ['ryanfranco', 'grantpatricia', 'brandtandrea', 'lorichen', 'mariokeller', 'ryanfranco', 'lorichen'],
    'rating': [5, 4, 5, 3, 4, 5, 5]
})

# Reseñas de texto
reviews_data = pd.DataFrame({
    'review_id': [1, 2, 3, 4],
    'book_id': [1, 2, 3, 5],
    'username': ['ryanfranco', 'lorichen', 'mariokeller', 'ryanfranco'],
    'text': ['Excellent book!', 'Good plot.', 'Amazing graphic novel.', 'Loved the sequel.']
})

# 3. Guardar los DataFrames en la base de datos SQLite local
authors_data.to_sql('authors', con=engine, if_exists='replace', index=False)
publishers_data.to_sql('publishers', con=engine, if_exists='replace', index=False)
books_data.to_sql('books', con=engine, if_exists='replace', index=False)
ratings_data.to_sql('ratings', con=engine, if_exists='replace', index=False)
reviews_data.to_sql('reviews', con=engine, if_exists='replace', index=False)

print("¡Base de datos local 'libros_local.db' creada y poblada con éxito!")
```

### Tarea 3.1: Número de libros publicados después del 1 de enero de 2000
El objetivo de esta consulta es medir qué tan actualizado está el catálogo de la plataforma.

```python
# Consulta SQL para contar libros
query_1 = """
SELECT COUNT(*) AS total_books
FROM books
WHERE publication_date > '2000-01-01';
"""

# Ejecutar la consulta en nuestro motor local
df_resultado_1 = pd.io.sql.read_sql(query_1, con=engine)
df_resultado_1
```


De la muestra evaluada, únicamente 2 títulos corresponden al nuevo milenio. Esto nos indica que el catálogo tiene un fuerte componente de literatura clásica o de finales del siglo XX.

### Tarea 3.2: Número de reseñas y calificación promedio para cada libro
El objetivo de negocio: Queremos saber qué libros atraen a más gente (volumen de reseñas) y cuáles son los mejor calificados (promedio de estrellas). Esto ayuda a identificar los "bestsellers" de la plataforma.

```python
# Consulta avanzada con CTEs para evitar duplicación de agregaciones
query_2 = """
WITH con_promedios AS (
    SELECT book_id, 
           AVG(rating) AS rating_avg
    FROM ratings
    GROUP BY book_id
),
con_reseñas AS (
    SELECT book_id, 
           COUNT(review_id) AS total_reviews
    FROM reviews
    GROUP BY book_id
)
SELECT b.book_id,
       b.title,
       COALESCE(p.rating_avg, 0) AS rating_avg,
       COALESCE(r.total_reviews, 0) AS total_reviews
FROM books b
LEFT JOIN con_promedios p ON b.book_id = p.book_id
LEFT JOIN con_reseñas r ON b.book_id = r.book_id;
"""

# Ejecutamos en la base local
df_resultado_2 = pd.io.sql.read_sql(query_2, con=engine)
df_resultado_2
```

El cruce de métricas cualitativas y cuantitativas nos muestra que *Dragonfly in Amber* lidera la satisfacción de la plataforma con una calificación de 5.0 estrellas, seguido de por *Outlander* con 4.5 estrellas. Mientras que *The Stranger* no muestra interacción.


### Tarea 3.3: Editorial con el mayor número de libros de más de 50 páginas

```python
# Consulta para encontrar la editorial líder en contenido relevante
query_3 = """
SELECT p.publisher,
       COUNT(b.book_id) AS total_books
FROM books b
JOIN publishers p ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY total_books DESC
LIMIT 1;
"""

# Ejecutamos en la base local
df_resultado_3 = pd.io.sql.read_sql(query_3, con=engine)
df_resultado_3
```

La editorial *Penguin Books* se consolida como el principal proveedor estratégico de contenido de valor extendido, con un total de 3 títulos en el catálogo.

### Tarea 3.4: Autor con la calificación promedio más alta

```python
# Consulta de tres tablas con filtro condicional HAVING
query_4 = """
SELECT a.author,
       AVG(r.rating) AS avg_rating,
       COUNT(r.rating) AS total_ratings
FROM books b
JOIN ratings r ON b.book_id = r.book_id
JOIN authors a ON b.author_id = a.author_id
GROUP BY a.author
HAVING COUNT(r.rating) >= 2
ORDER BY avg_rating DESC
LIMIT 1;
"""

# Ejecutamos en la base local
df_resultado_4 = pd.io.sql.read_sql(query_4, con=engine)
df_resultado_4
```

Al aplicar una regla de consistencia estadística (filtrar solo autores con un volumen de interacción representativo mediante `HAVING`), *Diana Gabaldon* sobresale como la autora más valorada por la comunidad, con un puntaje promedio de 4.75 estrellas con 4 calificaciones individuales.

### Tarea 3.5: Número promedio de reseñas de texto entre los usuarios más activos

```python
# Consulta avanzada para promediar la actividad de los Power Users
query_5 = """
WITH user_stats AS (
    SELECT r.username,
           COUNT(DISTINCT r.book_id) AS total_ratings,
           COUNT(DISTINCT rev.review_id) AS total_reviews
    FROM ratings r
    LEFT JOIN reviews rev ON r.username = rev.username
    GROUP BY r.username
    HAVING COUNT(DISTINCT r.book_id) > 1
)
SELECT AVG(total_reviews) AS avg_text_reviews
FROM user_stats;
"""

# Ejecutamos en la base local
df_resultado_5 = pd.io.sql.read_sql(query_5, con=engine)
df_resultado_5
```

El uso de subconsultas avanzadas y expresiones de tabla comunes (CTEs) nos permite aislar a los usuarios de alto consumo de la plataforma. Analizar cuántas reseñas detalladas escriben en promedio nuestros usuarios más activos es importante, ya que el contenido generado por el usuario (UGC) actúa como el principal motor orgánico para que otros clientes descubran nuevos libros.
